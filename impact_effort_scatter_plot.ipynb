# Impact v. effort scatter plot

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
impact_effort = pd.read_csv('impact_effort.csv')
impact_effort
```

```python
graphtitle = 'Impact v. Effort'
graphsubtitle = 'Symphony Processes'
graphylabel = 'Impact'
graphxlabel = 'Effort'
```

```python
plt.figure(figsize=(8,6), dpi=100)
ax = plt.subplot(111)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.tick_params(direction='out')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.scatter(impact_effort['effort'], impact_effort['impact'], color="blue", \
           linewidth=0, linestyle="-", s=100)
for i, txt in enumerate( impact_effort.process ):
    ax.annotate( txt, ( impact_effort.effort[i] + 1, impact_effort.impact[i] + 1 ) )
plt.xlim(xmin=0, xmax=100)    
plt.ylim(ymin=0, ymax=100)
# Plot horizontal reference line
plt.plot([0, 100], [50, 50], color="red", \
         linewidth=1.0, linestyle="-")
# Plot vertical reference line
plt.plot([50, 50], [0, 100], color="red", \
         linewidth=1.0, linestyle="-")
#plt.title(graphtitle + '\n' + graphsubtitle, fontweight="bold")
plt.title(graphtitle + '\n' + graphsubtitle, fontweight="bold")
plt.ylabel(graphylabel)
plt.xlabel(graphxlabel)
plt.savefig('impact_effort.svg', \
            format='svg');
plt.savefig('impact_effort.pdf', \
            format='pdf');
plt.savefig('impact_effort.png', \
            format='png');
plt.show()
```
